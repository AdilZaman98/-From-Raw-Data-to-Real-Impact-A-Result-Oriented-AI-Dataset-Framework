# **Outliers of the dataset:**

```python
# Step 1: Numeric columns select karo
numeric_cols = df.select_dtypes(include=['number']).columns
print("\nNumeric columns:", list(numeric_cols))

# Step 2: Outliers count karo (aapka original code)
print("\n=== OUTLIER DETECTION ===")
outlier_summary = []
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outlier_count = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
    if outlier_count > 0:
        print(f"Column '{col}': {outlier_count} outliers")
        outlier_summary.append({'Column': col, 'Outliers': outlier_count})

# Step 3: Data clean karo (aapka original code + improvements)
df_clean = df.copy()
print("\n=== CLEANING DATA ===")
for col in numeric_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    initial_rows = len(df_clean)
    df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]
    if len(df_clean) < initial_rows:
        print(f"{col}: {initial_rows - len(df_clean)} rows removed")

print(f"Original shape: {df.shape}")
print(f"Cleaned shape: {df_clean.shape}")

# Step 4-5: Scaled Boxplots (0-50 range) - DO GRAPHSCALED
plt.figure(figsize=(16, 12))
first_10_cols = numeric_cols[:10]

plt.subplot(2, 1, 1)
sns.boxplot(data=df[first_10_cols])
plt.ylim(0, 50)  # Scaling: 0 to 50
plt.xticks(rotation=45)
plt.title("Original Data: First 10 Columns (Scaled 0-50)")

plt.subplot(2, 1, 2)
sns.boxplot(data=df_clean[first_10_cols])
plt.ylim(0, 50)  # Scaling: 0 to 50
plt.xticks(rotation=45)
plt.title("Cleaned Data: First 10 Columns (Scaled 0-50)")

plt.tight_layout()
plt.show()

# Summary table
print("\n=== SUMMARY ===")
summary_df = pd.DataFrame(outlier_summary)
print(summary_df)
print("Data: df (original), df_clean (cleaned)")
```

```python
## 📦 Cell 1: Install Required Libraries
# Install all required libraries
!pip install numpy pandas matplotlib seaborn scikit-learn torch torchvision
!pip install transformers datasets pillow opencv-python
!pip install statsmodels prophet xgboost lightgbm
!pip install plotly ipywidgets tqdm
print('✅ All libraries installed successfully!')
## 🔧 Cell 2: Import Libraries & Setup
# ─── Standard Libraries ───────────────────────────────────────────────────────
import matplotlib.patches as mpatches
import warnings
import time
import os
import random
from datetime import datetime, timedelta
warnings.filterwarnings('ignore')
# ─── Machine Learning ─────────────────────────────────────────────────────────
from sklearn.datasets import load_iris, make_classification, make_blobs
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder, MinMaxScaler
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, IsolationForest
from sklearn.svm import SVC, OneClassSVM
from sklearn.neighbors import LocalOutlierFactor
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_auc_score, f1_score, mean_squared_error, r2_score
)
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
import xgboost as xgb

# ─── Deep Learning (PyTorch) ──────────────────────────────────────────────────
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import torchvision
import torchvision.transforms as transforms

# ─── NLP ──────────────────────────────────────────────────────────────────────
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification

# ─── Computer Vision ──────────────────────────────────────────────────────────
import cv2
from PIL import Image

# ─── Time Series ──────────────────────────────────────────────────────────────
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose

# ─── Plotting ─────────────────────────────────────────────────────────────────
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# ─── Configuration ────────────────────────────────────────────────────────────
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
torch.manual_seed(RANDOM_SEED)
random.seed(RANDOM_SEED)

DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# ─── Style ────────────────────────────────────────────────────────────────────
plt.style.use('dark_background')
sns.set_palette('husl')

print(f'✅ All imports successful!')
print(f'🔥 Device: {DEVICE}')
print(f'📅 Timestamp: {datetime.now().strftime("%Y-%m-%d %H:%M:%S")}')
## 🧩 Cell 3: Unified AI Framework — Master Class
class UnifiedAIFramework:
    """
    ╔══════════════════════════════════════════════════════════╗
    ║         UNIFIED AI FRAMEWORK — MASTER CONTROLLER        ║
    ║  ML | DL | NLP | Vision | Anomaly | TimeSeries          ║
    ╚══════════════════════════════════════════════════════════╝
    Integrates:
      - Machine Learning (RF, XGBoost, SVM, GBM)
      - Deep Learning (MLP, CNN via PyTorch)
      - NLP (Transformers: sentiment, classification)
      - Computer Vision (edge detection, feature extraction)
      - Anomaly Detection (Isolation Forest, LOF, One-Class SVM)
      - Time Series (ARIMA, LSTM forecasting)
    """

    def __init__(self, config=None):
        self.config = config or {
            'random_seed': 42,
            'test_size': 0.2,
            'n_estimators': 100,
            'max_depth': 6,
            'confidence_threshold': 0.7,
            'anomaly_contamination': 0.1,
        }
        self.device = DEVICE
        self.models = {}
        self.results = {}
        self.scaler = StandardScaler()
        print(f'🚀 UnifiedAIFramework initialized | Device: {self.device}')
        print(f'⚙️  Config: {self.config}')

    # ══════════════════════════════════════════════════════════
    # MODULE 1: DATA GENERATION & PREPROCESSING
    # ══════════════════════════════════════════════════════════
    def generate_multimodal_data(self):
        """Generate synthetic multi-modal dataset."""
        print('\n📊 [MODULE 1] Generating Multi-Modal Data...')

        # Tabular Classification Data
        X_cls, y_cls = make_classification(
            n_samples=2000, n_features=20, n_informative=12,
            n_redundant=4, n_classes=3, random_state=self.config['random_seed']
        )

        # Time Series Data (with trend + seasonality + noise)
        n_ts = 500
        dates = pd.date_range('2020-01-01', periods=n_ts, freq='D')
        trend = np.linspace(100, 200, n_ts)
        seasonality = 20 * np.sin(2 * np.pi * np.arange(n_ts) / 30)
        noise = np.random.normal(0, 5, n_ts)
        anomaly_idx = np.random.choice(n_ts, 20, replace=False)
        anomaly_signal = np.zeros(n_ts)
        anomaly_signal[anomaly_idx] = np.random.choice([-50, 50], 20)
        ts_values = trend + seasonality + noise + anomaly_signal
        ts_data = pd.DataFrame({'date': dates, 'value': ts_values})
        ts_data['is_anomaly'] = 0
        ts_data.loc[anomaly_idx, 'is_anomaly'] = 1

        # Text Data for NLP
        text_samples = [
            ('The AI model achieved outstanding accuracy on all benchmarks.', 'positive'),
            ('System failure detected in critical pipeline module.', 'negative'),
            ('Performance metrics remain stable within normal range.', 'neutral'),
            ('Anomaly detected: unauthorized access pattern identified.', 'negative'),
            ('Model training converged successfully with 98% accuracy.', 'positive'),
            ('Data quality issues found in 15% of the records.', 'negative'),
            ('Real-time inference latency improved by 40%.', 'positive'),
            ('Forecasting module predicts 12% revenue increase.', 'positive'),
            ('Critical error in anomaly detection subsystem.', 'negative'),
            ('All systems operational. No alerts detected.', 'neutral'),
        ] * 20  # replicate for volume
        text_df = pd.DataFrame(text_samples, columns=['text', 'sentiment'])

        # Image-like data (flattened for CNN)
        n_images = 600
        image_data = np.random.rand(n_images, 1, 28, 28).astype(np.float32)
        image_labels = np.random.randint(0, 5, n_images)

        self.data = {
            'classification': (X_cls, y_cls),
            'time_series': ts_data,
            'text': text_df,
            'image': (image_data, image_labels),
        }
        print(f'  ✅ Classification samples: {X_cls.shape}')
        print(f'  ✅ Time series points: {len(ts_data)} | Anomalies: {ts_data["is_anomaly"].sum()}')
        print(f'  ✅ Text samples: {len(text_df)}')
        print(f'  ✅ Image data shape: {image_data.shape}')
        return self.data

    def preprocess(self):
        """Preprocess all modalities."""
        print('\n🔧 [PREPROCESS] Scaling & Splitting...')
        X, y = self.data['classification']
        X_scaled = self.scaler.fit_transform(X)
        X_train, X_test, y_train, y_test = train_test_split(
            X_scaled, y, test_size=self.config['test_size'],
            random_state=self.config['random_seed'], stratify=y
        )
        self.splits = {
            'X_train': X_train, 'X_test': X_test,
            'y_train': y_train, 'y_test': y_test
        }
        print(f'  ✅ Train: {X_train.shape} | Test: {X_test.shape}')
        return self.splits

    # ══════════════════════════════════════════════════════════
    # MODULE 2: MACHINE LEARNING
    # ══════════════════════════════════════════════════════════
    def run_ml_module(self):
        """Train and evaluate multiple ML classifiers."""
        print('\n🤖 [MODULE 2] Machine Learning Classification...')
        X_tr, X_te = self.splits['X_train'], self.splits['X_test']
        y_tr, y_te = self.splits['y_train'], self.splits['y_test']

        classifiers = {
            'Random Forest':      RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1),
            'XGBoost':            xgb.XGBClassifier(n_estimators=100, use_label_encoder=False,
                                                     eval_metric='mlogloss', random_state=42, verbosity=0),
            'Gradient Boosting':  GradientBoostingClassifier(n_estimators=100, random_state=42),
            'SVM':                SVC(probability=True, random_state=42),
            'Logistic Regression': LogisticRegression(max_iter=500, random_state=42),
        }

        ml_results = {}
        for name, clf in classifiers.items():
            t0 = time.time()
            clf.fit(X_tr, y_tr)
            y_pred = clf.predict(X_te)
            y_prob = clf.predict_proba(X_te)
            acc = accuracy_score(y_te, y_pred)
            f1  = f1_score(y_te, y_pred, average='weighted')
            auc = roc_auc_score(y_te, y_prob, multi_class='ovr', average='weighted')
            elapsed = time.time() - t0
            ml_results[name] = {'accuracy': acc, 'f1': f1, 'auc': auc, 'time': elapsed}
            self.models[name] = clf
            print(f'  [{name:22s}] Acc={acc:.4f} | F1={f1:.4f} | AUC={auc:.4f} | ⏱ {elapsed:.2f}s')

        self.results['ml'] = ml_results
        best = max(ml_results, key=lambda k: ml_results[k]['accuracy'])
        print(f'\n  🏆 Best ML Model: {best} (Acc={ml_results[best]["accuracy"]:.4f})')
        return ml_results

    # ══════════════════════════════════════════════════════════
    # MODULE 3: DEEP LEARNING (MLP + CNN)
    # ══════════════════════════════════════════════════════════
    def run_dl_module(self, epochs=15):
        """Train MLP and CNN deep learning models."""
        print('\n🧠 [MODULE 3] Deep Learning (MLP + CNN)...')

        # ── MLP for tabular classification ──
        class MLP(nn.Module):
            def __init__(self, in_dim, hidden_dims, out_dim, dropout=0.3):
                super().__init__()
                layers = []
                prev = in_dim
                for h in hidden_dims:
                    layers += [nn.Linear(prev, h), nn.BatchNorm1d(h), nn.ReLU(), nn.Dropout(dropout)]
                    prev = h
                layers.append(nn.Linear(prev, out_dim))
                self.net = nn.Sequential(*layers)
            def forward(self, x): return self.net(x)

        # ── CNN for image-like data ──
        class CNN(nn.Module):
            def __init__(self, n_classes=5):
                super().__init__()
                self.features = nn.Sequential(
                    nn.Conv2d(1, 32, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
                    nn.Conv2d(32, 64, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
                    nn.Dropout2d(0.25)
                )
                self.classifier = nn.Sequential(
                    nn.Flatten(),
                    nn.Linear(64 * 7 * 7, 128), nn.ReLU(), nn.Dropout(0.5),
                    nn.Linear(128, n_classes)
                )
            def forward(self, x): return self.classifier(self.features(x))

        def train_model(model, loader, criterion, optimizer, n_epochs, label):
            model.train()
            history = []
            for ep in range(n_epochs):
                total_loss, correct, total = 0, 0, 0
                for X_batch, y_batch in loader:
                    X_batch, y_batch = X_batch.to(self.device), y_batch.to(self.device)
                    optimizer.zero_grad()
                    out = model(X_batch)
                    loss = criterion(out, y_batch)
                    loss.backward()
                    optimizer.step()
                    total_loss += loss.item()
                    correct += (out.argmax(1) == y_batch).sum().item()
                    total += len(y_batch)
                acc = correct / total
                history.append({'epoch': ep+1, 'loss': total_loss/len(loader), 'acc': acc})
                if (ep+1) % 5 == 0:
                    print(f'  [{label}] Epoch {ep+1:3d}/{n_epochs} | Loss={total_loss/len(loader):.4f} | Acc={acc:.4f}')
            return history

        # ── Prepare MLP data ──
        X_tr = torch.FloatTensor(self.splits['X_train'])
        y_tr = torch.LongTensor(self.splits['y_train'])
        mlp_loader = DataLoader(TensorDataset(X_tr, y_tr), batch_size=64, shuffle=True)
        mlp = MLP(X_tr.shape[1], [256, 128, 64], 3).to(self.device)
        mlp_opt = optim.Adam(mlp.parameters(), lr=1e-3, weight_decay=1e-4)
        mlp_hist = train_model(mlp, mlp_loader, nn.CrossEntropyLoss(), mlp_opt, epochs, 'MLP')

        # ── Prepare CNN data ──
        imgs, img_labels = self.data['image']
        imgs_t = torch.FloatTensor(imgs)
        lbls_t = torch.LongTensor(img_labels)
        cnn_loader = DataLoader(TensorDataset(imgs_t, lbls_t), batch_size=32, shuffle=True)
        cnn = CNN(n_classes=5).to(self.device)
        cnn_opt = optim.Adam(cnn.parameters(), lr=1e-3)
        cnn_hist = train_model(cnn, cnn_loader, nn.CrossEntropyLoss(), cnn_opt, epochs, 'CNN')

        self.models['MLP'] = mlp
        self.models['CNN'] = cnn
        self.results['dl'] = {'mlp_history': mlp_hist, 'cnn_history': cnn_hist}
        print(f'\n  ✅ MLP Final Acc: {mlp_hist[-1]["acc"]:.4f}')
        print(f'  ✅ CNN Final Acc: {cnn_hist[-1]["acc"]:.4f}')
        return self.results['dl']

    # ══════════════════════════════════════════════════════════
    # MODULE 4: NLP — TEXT UNDERSTANDING
    # ══════════════════════════════════════════════════════════
    def run_nlp_module(self):
        """Perform sentiment analysis and text classification via Transformers."""
        print('\n📝 [MODULE 4] NLP — Text Understanding...')
        texts = self.data['text']['text'].tolist()[:30]  # sample for speed

        # Sentiment Analysis
        print('  Loading sentiment pipeline (distilbert)...')
        sentiment_pipe = pipeline(
            'sentiment-analysis',
            model='distilbert-base-uncased-finetuned-sst-2-english',
            truncation=True, max_length=128
        )
        sentiments = sentiment_pipe(texts)

        nlp_df = pd.DataFrame({
            'text': texts,
            'sentiment_label': [s['label'] for s in sentiments],
            'confidence': [s['score'] for s in sentiments]
        })

        avg_conf = nlp_df['confidence'].mean()
        dist = nlp_df['sentiment_label'].value_counts().to_dict()

        print(f'  ✅ Analyzed {len(texts)} samples')
        print(f'  📊 Distribution: {dist}')
        print(f'  🎯 Avg Confidence: {avg_conf:.4f}')

        # Zero-shot classification
        print('  Running zero-shot classification...')
        zs_pipe = pipeline('zero-shot-classification',
                           model='facebook/bart-large-mnli')
        candidate_labels = ['anomaly alert', 'system health', 'performance report', 'security incident']
        zs_results = []
        for t in texts[:10]:
            res = zs_pipe(t, candidate_labels)
            zs_results.append({
                'text': t[:60] + '...',
                'top_label': res['labels'][0],
                'top_score': res['scores'][0]
            })

        zs_df = pd.DataFrame(zs_results)
        print(f'\n  🔍 Zero-Shot Results (top 5):')
        print(zs_df.head().to_string(index=False))

        self.results['nlp'] = {'sentiment_df': nlp_df, 'zeroshot_df': zs_df}
        return self.results['nlp']

    # ══════════════════════════════════════════════════════════
    # MODULE 5: COMPUTER VISION
    # ══════════════════════════════════════════════════════════
    def run_vision_module(self):
        """Perform visual feature extraction and edge detection."""
        print('\n👁️ [MODULE 5] Computer Vision — Visual Recognition...')
        imgs, _ = self.data['image']
        sample = (imgs[0, 0] * 255).astype(np.uint8)

        # Edge detection
        edges = cv2.Canny(sample, 50, 150)
        blurred = cv2.GaussianBlur(sample, (5, 5), 0)
        _, thresh = cv2.threshold(sample, 127, 255, cv2.THRESH_BINARY)

        # Feature stats
        features = {
            'mean_intensity': float(np.mean(sample)),
            'std_intensity':  float(np.std(sample)),
            'edge_density':   float(np.sum(edges > 0) / edges.size),
            'bright_ratio':   float(np.sum(sample > 127) / sample.size),
        }

        # PCA for dimensionality reduction on image batch
        flat_imgs = imgs.reshape(len(imgs), -1)
        pca = PCA(n_components=50, random_state=42)
        img_pca = pca.fit_transform(flat_imgs)
        explained = pca.explained_variance_ratio_.sum()

        print(f'  ✅ Edge detection complete')
        print(f'  📊 Image features: {features}')
        print(f'  🔬 PCA (50 components) variance explained: {explained:.4f}')

        self.results['vision'] = {
            'sample': sample, 'edges': edges,
            'blurred': blurred, 'thresh': thresh,
            'features': features, 'pca_data': img_pca,
            'pca_variance': explained
        }
        return self.results['vision']

    # ══════════════════════════════════════════════════════════
    # MODULE 6: ANOMALY DETECTION
    # ══════════════════════════════════════════════════════════
    def run_anomaly_module(self):
        """Multi-method anomaly detection."""
        print('\n🚨 [MODULE 6] Anomaly Detection...')
        X_tr = self.splits['X_train']
        ts = self.data['time_series']

        # ── Statistical: Z-score on time series ──
        z_scores = np.abs((ts['value'] - ts['value'].mean()) / ts['value'].std())
        ts['zscore_anomaly'] = (z_scores > 3).astype(int)

        # ── Isolation Forest ──
        iso = IsolationForest(contamination=self.config['anomaly_contamination'],
                              random_state=42, n_jobs=-1)
        iso_labels = iso.fit_predict(X_tr)
        iso_anomaly_pct = (iso_labels == -1).mean() * 100

        # ── Local Outlier Factor ──
        lof = LocalOutlierFactor(n_neighbors=20, contamination=self.config['anomaly_contamination'])
        lof_labels = lof.fit_predict(X_tr)
        lof_anomaly_pct = (lof_labels == -1).mean() * 100

        # ── One-Class SVM ──
        oc_svm = OneClassSVM(nu=0.1, kernel='rbf', gamma='scale')
        oc_labels = oc_svm.fit_predict(X_tr[:500])  # subset for speed
        svm_anomaly_pct = (oc_labels == -1).mean() * 100

        # ── LSTM Autoencoder for TS anomaly ──
        class LSTMAutoencoder(nn.Module):
            def __init__(self, seq_len, n_features, hidden=64):
                super().__init__()
                self.enc = nn.LSTM(n_features, hidden, batch_first=True)
                self.dec = nn.LSTM(hidden, n_features, batch_first=True)
            def forward(self, x):
                _, (h, _) = self.enc(x)
                h_rep = h.permute(1, 0, 2).repeat(1, x.size(1), 1)
                out, _ = self.dec(h_rep)
                return out

        ts_vals = ts['value'].values.astype(np.float32)
        ts_norm = (ts_vals - ts_vals.mean()) / ts_vals.std()
        window, stride = 30, 1
        windows = np.array([ts_norm[i:i+window] for i in range(0, len(ts_norm)-window, stride)])
        ts_tensor = torch.FloatTensor(windows).unsqueeze(-1).to(self.device)

        lstm_ae = LSTMAutoencoder(window, 1).to(self.device)
        ae_opt = optim.Adam(lstm_ae.parameters(), lr=1e-3)
        lstm_ae.train()
        for ep in range(10):
            ae_opt.zero_grad()
            recon = lstm_ae(ts_tensor)
            loss = nn.MSELoss()(recon, ts_tensor)
            loss.backward()
            ae_opt.step()

        lstm_ae.eval()
        with torch.no_grad():
            recon = lstm_ae(ts_tensor)
        recon_errors = ((ts_tensor - recon)**2).mean(dim=(1,2)).cpu().numpy()
        lstm_threshold = np.percentile(recon_errors, 90)
        lstm_anomalies = (recon_errors > lstm_threshold).sum()

        anomaly_results = {
            'zscore_anomalies':   int(ts['zscore_anomaly'].sum()),
            'isolation_forest_%': round(iso_anomaly_pct, 2),
            'lof_%':              round(lof_anomaly_pct, 2),
            'ocsvm_%':            round(svm_anomaly_pct, 2),
            'lstm_ae_anomalies':  int(lstm_anomalies),
            'ts_with_labels':     ts,
            'recon_errors':       recon_errors,
            'lstm_threshold':     float(lstm_threshold)
        }

        print(f'  ✅ Z-Score Anomalies (TS): {anomaly_results["zscore_anomalies"]}')
        print(f'  ✅ Isolation Forest: {anomaly_results["isolation_forest_%"]}% anomalous')
        print(f'  ✅ LOF:              {anomaly_results["lof_%"]}% anomalous')
        print(f'  ✅ One-Class SVM:    {anomaly_results["ocsvm_%"]}% anomalous')
        print(f'  ✅ LSTM-AE Anomalies: {anomaly_results["lstm_ae_anomalies"]} windows')

        self.results['anomaly'] = anomaly_results
        return anomaly_results

    # ══════════════════════════════════════════════════════════
    # MODULE 7: TIME SERIES FORECASTING
    # ══════════════════════════════════════════════════════════
    def run_timeseries_module(self, forecast_steps=30):
        """ARIMA + LSTM forecasting."""
        print('\n📈 [MODULE 7] Time Series Forecasting...')
        ts = self.data['time_series'].copy()
        ts.set_index('date', inplace=True)

        # ── ARIMA Forecasting ──
        train_ts = ts['value'][:-forecast_steps]
        test_ts  = ts['value'][-forecast_steps:]
        try:
            arima_model = ARIMA(train_ts, order=(2, 1, 2))
            arima_fit = arima_model.fit()
            arima_forecast = arima_fit.forecast(steps=forecast_steps)
            arima_rmse = np.sqrt(mean_squared_error(test_ts, arima_forecast))
            arima_r2   = r2_score(test_ts, arima_forecast)
            print(f'  ✅ ARIMA(2,1,2) RMSE: {arima_rmse:.4f} | R²: {arima_r2:.4f}')
        except Exception as e:
            arima_forecast = pd.Series(np.random.randn(forecast_steps))
            arima_rmse = 999
            print(f'  ⚠️  ARIMA fallback: {e}')

        # ── LSTM Forecasting ──
        class LSTMForecaster(nn.Module):
            def __init__(self, in_dim=1, hidden=128, n_layers=2, out_dim=1):
                super().__init__()
                self.lstm = nn.LSTM(in_dim, hidden, n_layers, batch_first=True, dropout=0.2)
                self.fc = nn.Linear(hidden, out_dim)
            def forward(self, x):
                out, _ = self.lstm(x)
                return self.fc(out[:, -1, :])

        vals = ts['value'].values.astype(np.float32)
        scaler_ts = MinMaxScaler()
        vals_sc = scaler_ts.fit_transform(vals.reshape(-1,1)).flatten()

        SEQ = 20
        Xs = np.array([vals_sc[i:i+SEQ] for i in range(len(vals_sc)-SEQ-forecast_steps)])
        ys = np.array([vals_sc[i+SEQ] for i in range(len(vals_sc)-SEQ-forecast_steps)])
        X_tt = torch.FloatTensor(Xs).unsqueeze(-1).to(self.device)
        y_tt = torch.FloatTensor(ys).unsqueeze(-1).to(self.device)
        ds = TensorDataset(X_tt, y_tt)
        dl = DataLoader(ds, batch_size=32, shuffle=True)

        lstm_fc = LSTMForecaster().to(self.device)
        opt = optim.Adam(lstm_fc.parameters(), lr=1e-3)
        crit = nn.MSELoss()
        for ep in range(20):
            lstm_fc.train()
            for xb, yb in dl:
                opt.zero_grad()
                crit(lstm_fc(xb), yb).backward()
                opt.step()

        lstm_fc.eval()
        # Multi-step forecast
        context = torch.FloatTensor(vals_sc[-SEQ:]).unsqueeze(0).unsqueeze(-1).to(self.device)
        lstm_preds = []
        with torch.no_grad():
            for _ in range(forecast_steps):
                p = lstm_fc(context).item()
                lstm_preds.append(p)
                new = torch.FloatTensor([[[p]]]).to(self.device)
                context = torch.cat([context[:, 1:, :], new], dim=1)
        lstm_preds_inv = scaler_ts.inverse_transform(
            np.array(lstm_preds).reshape(-1,1)).flatten()
        lstm_rmse = np.sqrt(mean_squared_error(test_ts.values, lstm_preds_inv))
        lstm_r2   = r2_score(test_ts.values, lstm_preds_inv)

        print(f'  ✅ LSTM Forecaster RMSE: {lstm_rmse:.4f} | R²: {lstm_r2:.4f}')

        forecast_dates = pd.date_range(
            ts.index[-forecast_steps], periods=forecast_steps, freq='D')

        self.results['timeseries'] = {
            'historical': ts['value'],
            'test_actual': test_ts,
            'arima_forecast': arima_forecast,
            'lstm_forecast': pd.Series(lstm_preds_inv, index=forecast_dates),
            'arima_rmse': arima_rmse,
            'lstm_rmse': lstm_rmse,
            'arima_r2': arima_r2,
            'lstm_r2': lstm_r2,
        }
        return self.results['timeseries']

    # ══════════════════════════════════════════════════════════
    # MODULE 8: CONFIDENCE SCORING & RISK ASSESSMENT
    # ══════════════════════════════════════════════════════════
    def compute_confidence_risk(self):
        """Compute system-wide confidence and risk scores."""
        print('\n🛡️ [MODULE 8] Confidence & Risk Assessment...')
        ml = self.results.get('ml', {})
        dl = self.results.get('dl', {})
        ts = self.results.get('timeseries', {})
        anomaly = self.results.get('anomaly', {})

        ml_best_acc = max([v['accuracy'] for v in ml.values()], default=0)
        mlp_acc = dl.get('mlp_history', [{}])[-1].get('acc', 0) if dl else 0
        cnn_acc = dl.get('cnn_history', [{}])[-1].get('acc', 0) if dl else 0

        arima_norm = max(0, 1 - ts.get('arima_rmse', 999) / 50) if ts else 0
        lstm_norm  = max(0, 1 - ts.get('lstm_rmse', 999) / 50)  if ts else 0

        iso_pct = anomaly.get('isolation_forest_%', 50) / 100
        lof_pct = anomaly.get('lof_%', 50) / 100

        # Weighted confidence
        confidence = (
            0.30 * ml_best_acc +
            0.20 * mlp_acc +
            0.15 * cnn_acc +
            0.20 * (arima_norm + lstm_norm) / 2 +
            0.15 * (1 - (iso_pct + lof_pct) / 2)
        )

        # Risk score (higher anomaly = higher risk)
        risk = (
            0.40 * ((iso_pct + lof_pct) / 2) +
            0.30 * (1 - ml_best_acc) +
            0.30 * (1 - (arima_norm + lstm_norm) / 2)
        )

        risk_level = 'LOW' if risk < 0.3 else ('MEDIUM' if risk < 0.6 else 'HIGH')
        decision = 'AUTOMATED_SAFE' if (confidence > self.config['confidence_threshold'] and risk < 0.3) else \
                   'HUMAN_REVIEW'   if risk < 0.6 else 'HALT_AND_AUDIT'

        self.results['risk'] = {
            'confidence': round(confidence, 4),
            'risk_score':  round(risk, 4),
            'risk_level':  risk_level,
            'decision':    decision,
            'ml_best_acc': round(ml_best_acc, 4),
            'mlp_acc':     round(mlp_acc, 4),
            'cnn_acc':     round(cnn_acc, 4),
        }
        print(f'  🎯 Confidence Score:  {confidence:.4f}')
        print(f'  ⚠️  Risk Score:        {risk:.4f} [{risk_level}]')
        print(f'  🤖 Decision:          {decision}')
        return self.results['risk']

    # ══════════════════════════════════════════════════════════
    # MODULE 9: UNIFIED DASHBOARD VISUALIZATION
    # ══════════════════════════════════════════════════════════
    def visualize_all(self):
        """Generate a comprehensive visualization dashboard."""
        print('\n📊 [MODULE 9] Generating Unified Dashboard...')
        fig = plt.figure(figsize=(24, 28))
        fig.patch.set_facecolor('#0d1117')
        colors = ['#00d4ff', '#ff6b6b', '#51cf66', '#ffd43b', '#cc5de8', '#ff8cc8']

        def ax_style(ax, title):
            ax.set_facecolor('#161b22')
            ax.set_title(title, color='white', fontsize=11, fontweight='bold', pad=8)
            ax.tick_params(colors='#8b949e')
            for spine in ax.spines.values(): spine.set_color('#30363d')
            ax.xaxis.label.set_color('#8b949e')
            ax.yaxis.label.set_color('#8b949e')

        # ── 1. ML Model Comparison ──
        ax1 = fig.add_subplot(4, 4, 1)
        if 'ml' in self.results:
            names = list(self.results['ml'].keys())
            accs  = [v['accuracy'] for v in self.results['ml'].values()]
            bars  = ax1.bar(range(len(names)), accs, color=colors[:len(names)])
            ax1.set_xticks(range(len(names)))
            ax1.set_xticklabels([n.split()[0] for n in names], rotation=30, fontsize=8, color='#8b949e')
            ax1.set_ylim(0, 1)
            for b, a in zip(bars, accs):
                ax1.text(b.get_x() + b.get_width()/2, a + 0.01, f'{a:.3f}',
                         ha='center', va='bottom', color='white', fontsize=8)
        ax_style(ax1, '🤖 ML Accuracy Comparison')

        # ── 2. ML Metrics Radar (as bar) ──
        ax2 = fig.add_subplot(4, 4, 2)
        if 'ml' in self.results:
            metrics = ['accuracy', 'f1', 'auc']
            x = np.arange(len(metrics))
            width = 0.15
            for i, (name, vals) in enumerate(self.results['ml'].items()):
                vals_m = [vals[m] for m in metrics]
                ax2.bar(x + i*width, vals_m, width, label=name.split()[0], color=colors[i % len(colors)], alpha=0.8)
            ax2.set_xticks(x + width*2)
            ax2.set_xticklabels(metrics, color='#8b949e', fontsize=9)
            ax2.legend(fontsize=6, loc='lower right', facecolor='#161b22', labelcolor='white')
            ax2.set_ylim(0, 1.1)
        ax_style(ax2, '📊 ML Multi-Metric Comparison')

        # ── 3. DL Training Curves ──
        ax3 = fig.add_subplot(4, 4, 3)
        if 'dl' in self.results:
            for key, col, lbl in [('mlp_history', colors[0], 'MLP'), ('cnn_history', colors[1], 'CNN')]:
                h = self.results['dl'][key]
                eps = [e['epoch'] for e in h]
                acc = [e['acc'] for e in h]
                ax3.plot(eps, acc, color=col, label=lbl, linewidth=2)
            ax3.legend(fontsize=9, facecolor='#161b22', labelcolor='white')
            ax3.set_xlabel('Epoch')
            ax3.set_ylabel('Accuracy')
        ax_style(ax3, '🧠 DL Training Accuracy')

        # ── 4. DL Loss Curves ──
        ax4 = fig.add_subplot(4, 4, 4)
        if 'dl' in self.results:
            for key, col, lbl in [('mlp_history', colors[2], 'MLP'), ('cnn_history', colors[3], 'CNN')]:
                h = self.results['dl'][key]
                eps = [e['epoch'] for e in h]
                loss = [e['loss'] for e in h]
                ax4.plot(eps, loss, color=col, label=lbl, linewidth=2)
            ax4.legend(fontsize=9, facecolor='#161b22', labelcolor='white')
            ax4.set_xlabel('Epoch')
            ax4.set_ylabel('Loss')
        ax_style(ax4, '📉 DL Loss Curves')

        # ── 5. Time Series + Anomalies ──
        ax5 = fig.add_subplot(4, 4, (5, 6))
        ts_data = self.data['time_series']
        ax5.plot(ts_data['date'], ts_data['value'], color=colors[0], linewidth=1, label='Signal', alpha=0.9)
        anoms = ts_data[ts_data['is_anomaly'] == 1]
        ax5.scatter(anoms['date'], anoms['value'], color=colors[1], s=60, zorder=5, label='True Anomaly')
        if 'anomaly' in self.results:
            z_anoms = ts_data[ts_data.get('zscore_anomaly', pd.Series(0, index=ts_data.index)) == 1]
            if hasattr(self.results['anomaly']['ts_with_labels'], 'index'):
                ts_labeled = self.results['anomaly']['ts_with_labels'].reset_index()
                z_det = ts_labeled[ts_labeled['zscore_anomaly'] == 1]
                ax5.scatter(z_det['date'], z_det['value'], color=colors[3], s=40,
                            marker='^', zorder=4, label='Z-Score Detected')
        ax5.legend(fontsize=8, facecolor='#161b22', labelcolor='white')
        ax_style(ax5, '📈 Time Series with Anomaly Detection')

        # ── 6. ARIMA & LSTM Forecast ──
        ax6 = fig.add_subplot(4, 4, (7, 8))
        if 'timeseries' in self.results:
            ts_r = self.results['timeseries']
            hist = ts_r['historical']
            ax6.plot(hist.index[-100:], hist.values[-100:], color=colors[0], lw=2, label='Historical')
            test = ts_r['test_actual']
            ax6.plot(test.index, test.values, color='white', lw=2, linestyle='--', label='Actual')
            arima_f = ts_r['arima_forecast']
            ax6.plot(test.index, arima_f.values[:len(test)], color=colors[2], lw=2, label=f'ARIMA (RMSE={ts_r["arima_rmse"]:.2f})')
            lstm_f = ts_r['lstm_forecast']
            ax6.plot(test.index, lstm_f.values[:len(test)], color=colors[3], lw=2, label=f'LSTM (RMSE={ts_r["lstm_rmse"]:.2f})')
            ax6.legend(fontsize=8, facecolor='#161b22', labelcolor='white')
        ax_style(ax6, '🔮 ARIMA vs LSTM Forecasting')

        # ── 7. LSTM Reconstruction Error ──
        ax7 = fig.add_subplot(4, 4, 9)
        if 'anomaly' in self.results and 'recon_errors' in self.results['anomaly']:
            errs = self.results['anomaly']['recon_errors']
            thresh = self.results['anomaly']['lstm_threshold']
            ax7.plot(errs, color=colors[0], lw=1, alpha=0.8)
            ax7.axhline(thresh, color=colors[1], lw=2, linestyle='--', label=f'Threshold={thresh:.4f}')
            ax7.fill_between(range(len(errs)), thresh, np.where(errs > thresh, errs, thresh),
                             color=colors[1], alpha=0.3)
            ax7.legend(fontsize=7, facecolor='#161b22', labelcolor='white')
        ax_style(ax7, '🔬 LSTM-AE Reconstruction Error')

        # ── 8. Anomaly Methods Comparison ──
        ax8 = fig.add_subplot(4, 4, 10)
        if 'anomaly' in self.results:
            methods = ['Z-Score', 'Isolation\nForest', 'LOF', 'One-Class\nSVM', 'LSTM-AE']
            a = self.results['anomaly']
            vals_a = [
                a['zscore_anomalies'],
                a['isolation_forest_%'],
                a['lof_%'],
                a['ocsvm_%'],
                a['lstm_ae_anomalies']
            ]
            bars = ax8.bar(methods, vals_a, color=colors[:5])
            for b, v in zip(bars, vals_a):
                ax8.text(b.get_x() + b.get_width()/2, v + 0.3, f'{v:.1f}',
                         ha='center', va='bottom', color='white', fontsize=8)
            ax8.set_xticklabels(methods, fontsize=7, color='#8b949e')
        ax_style(ax8, '🚨 Anomaly Detection Methods')

        # ── 9. Computer Vision ──
        if 'vision' in self.results:
            for i, (key, title) in enumerate([
                ('sample', 'Original'), ('edges', 'Edges'), ('blurred', 'Blurred'), ('thresh', 'Threshold')
            ]):
                ax_v = fig.add_subplot(4, 4, 11 + i)
                img = self.results['vision'][key]
                ax_v.imshow(img, cmap='plasma' if key == 'sample' else 'gray')
                ax_v.axis('off')
                ax_v.set_title(f'👁️ {title}', color='white', fontsize=9, fontweight='bold')
                ax_v.set_facecolor('#161b22')

        # ── 10. Confidence & Risk Gauge ──
        ax10 = fig.add_subplot(4, 4, 15)
        if 'risk' in self.results:
            r = self.results['risk']
            labels = ['Confidence', 'Risk']
            values = [r['confidence'], r['risk_score']]
            bar_colors = [colors[2], colors[1]]
            ax10.barh(labels, values, color=bar_colors, height=0.4)
            for i, (l, v) in enumerate(zip(labels, values)):
                ax10.text(v + 0.01, i, f'{v:.4f}', va='center', color='white', fontsize=10)
            ax10.set_xlim(0, 1.2)
            ax10.axvline(self.config['confidence_threshold'], color='yellow', lw=2, linestyle='--', alpha=0.7)
        ax_style(ax10, f'🛡️ Risk: {self.results["risk"]["risk_level"] if "risk" in self.results else "N/A"}')

        # ── 11. Decision Summary ──
        ax11 = fig.add_subplot(4, 4, 16)
        ax11.set_facecolor('#161b22')
        ax11.axis('off')
        if 'risk' in self.results:
            r = self.results['risk']
            summary_text = [
                f"UNIFIED AI FRAMEWORK",
                f"─────────────────────",
                f"ML Best Acc:  {r['ml_best_acc']:.4f}",
                f"MLP Acc:      {r['mlp_acc']:.4f}",
                f"CNN Acc:      {r['cnn_acc']:.4f}",
                f"Confidence:   {r['confidence']:.4f}",
                f"Risk Score:   {r['risk_score']:.4f}",
                f"Risk Level:   {r['risk_level']}",
                f"─────────────────────",
                f"DECISION:",
                f"{r['decision']}",
            ]
            for i, line in enumerate(summary_text):
                color = colors[2] if 'SAFE' in line else colors[1] if 'HALT' in line else \
                        'yellow' if 'REVIEW' in line else 'white'
                ax11.text(0.05, 0.95 - i * 0.09, line, color=color,
                          fontsize=8.5, fontfamily='monospace', transform=ax11.transAxes, va='top')
        ax11.set_title('📋 Decision Summary', color='white', fontsize=11, fontweight='bold')

        plt.suptitle('🤖 UNIFIED AI FRAMEWORK — COMPLETE DASHBOARD',
                     fontsize=16, fontweight='bold', color='white', y=1.01)
        plt.tight_layout()
        plt.savefig('unified_ai_dashboard.png', dpi=150, bbox_inches='tight',
                    facecolor='#0d1117')
        plt.show()
        print('  ✅ Dashboard saved as unified_ai_dashboard.png')

    # ══════════════════════════════════════════════════════════
    # MASTER RUNNER
    # ══════════════════════════════════════════════════════════
    def run_all(self, skip_nlp=False):
        """Run the complete unified framework pipeline."""
        print('='*60)
        print('  UNIFIED AI FRAMEWORK — FULL PIPELINE')
        print('='*60)
        t_start = time.time()
        self.generate_multimodal_data()
        self.preprocess()
        self.run_ml_module()
        self.run_dl_module(epochs=15)
        if not skip_nlp:
            self.run_nlp_module()
        self.run_vision_module()
        self.run_anomaly_module()
        self.run_timeseries_module()
        self.compute_confidence_risk()
        self.visualize_all()
        total = time.time() - t_start
        print(f'\n✅ Pipeline complete in {total:.1f}s')
        return self.results

print('✅ UnifiedAIFramework class defined!')
## 🚀 Cell 4: Run the Complete Pipeline
# Initialize and run the framework
# skip_nlp=True for faster execution (NLP models can take time to download)
framework = UnifiedAIFramework()
results = framework.run_all(skip_nlp=True)  # Set skip_nlp=False to enable full NLP
```
